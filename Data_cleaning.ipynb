## Data cleaning

Data cleaning is the process of fixing or removing incorrect,corrupted,incorrectly formatted duplicate incomplete data within a data set.

```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('C:\\Users\\User\\Desktop\\data_science bootcamp\\Dataset\\modified_titanic_data.csv')
```

```python
titanic.head()
```

```python
# df = df.drop(columns="unnamed:13")
# or
titanic.drop(columns="Unnamed: 13",inplace = True)
titanic.head()
```

```python
##False() mark all duplicates as True

titanic[titanic.duplicated(keep=False)]
```

```python
## The len() python method retuns the length of a list,string,dictionary,or any other iterable data format in 
## python
len(titanic)
```

```python
## First () mark dulicates as True except for the first occurence
titanic = titanic.drop_duplicates(keep="first")
titanic
len(titanic)
```

```python
titanic[titanic.duplicated(keep=False)]
```
