### Load necessary lines

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)

plt.title('Our First Graph!',fontdict={'fontname' : 'Comic Sans Ms','fontsize' : 20})
plt.xlabel('x Axis')
plt.ylabel('y Axis')

plt.show()
```

### Bar chart

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

bars[0].set_hatch['/']
bars[1].set_hatch['O']
bars[2].set_hatch['*']

plt.figure(figsize=(6,4))

plt.show()
```
