### Eliminate punctuation and stop words from the sentence

```python
#import required library
import string
from nltk.corpus import stopwords
```

```python
#view first ten stopwords present in the english corpus
stopwords.words('english')[0:10]
```

```python
#create a text sentence
test_sentence = 'This is my first test string. Wow!! we are doing just fine'
```

```python
#eliminate the punctuation in form of characters and print them
no_punctuation = [char for char in test_sentence if char not in string.punctuation]
no_punctuation
```

```python
#now eliminate the punctuation and print them as a whole sentence
no_punctuation =''.join(no_punctuation)
no_punctuation
```

#split each words present in the new sentence
no_punctuation.split()

```python
# now eliminate stopwords
clean_sentence = [word for word in no_punctuation.split() if word.lower() not in stopwords.words('english')]
```

```python
#print the final cleaned sentence
clean_sentence
```

###### How to process documents using bag of words

```python
#import required libraries
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#instantiate the vectorizer
vectorizer =CountVectorizer()
```

```python
#create 3 documents
document1 = 'Hi How are you'
document2 = 'today is a very very very pleasant day and we can have some fun fun fun'
document3 = 'This was an amazing experience'
```

```python
#put them together
listofdocuments = [document1,document2,document3]
```

```python
#fit them as a bag of words
bag_of_words = vectorizer.fit(listofdocuments)
```

```python
#check bag of words
bag_of_words
```

```python
#Apply transform method
bag_of_words = vectorizer.transform(listofdocuments)
```

```python
#print bag of words
print (bag_of_words)
```

```python
#verify the vocabulary for repeated words
print (vectorizer.vocabulary_.get('very'))
print (vectorizer.vocabulary_.get('fun'))
```
